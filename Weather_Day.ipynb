```python
# import packages
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import datetime
import os
```

### 建立路徑及時間

```python
# set dir 
today = str(datetime.date.today())
cwb_data = "cwb_weather_data"
if not os.path.exists(cwb_data):
    os.mkdir(cwb_data)
```

### 連接api，並下載資料

```python
import urllib.request
import zipfile 
res ="http://opendata.cwb.gov.tw/govdownload?dataid=F-C0032-001&authorizationkey=rdec-key-123-45678-011121314"
urllib.request.urlretrieve(res,"F-C0032-001.xml")
filename = 'F-C0032-001.xml'
f= open('F-C0032-001.xml',"r")
```

### 整理資料

```python
CITY = []
DAY = []
TIME = []
WX = []
MAXT = []
MINT = []
CI = []
PoP = []
get_day = []

data = open('F-C0032-001.xml',"r", encoding='utf-8').read()
soup = BeautifulSoup(data,"xml")
a = soup.find_all("location")

for i in range(0,len(a)):
    location = a[i]
    City = location.find_all("locationName")[0].text
    CITY.append(City)
    time = location.startTime.text.split("T")
    DAY.append(time[0])
    TIME.append(time[1].split("+")[0])
    get_day.append(today)
    WX.append(location.find_all("parameterName")[0].text)
    MAXT.append(location.find_all("parameterName")[3].text)
    MINT.append(location.find_all("parameterName")[6].text)
    CI.append(location.find_all("parameterName")[9].text)
    PoP.append(location.find_all("parameterName")[12].text)
```

```python
data = {"CITY":CITY,"DAY":DAY,"TIME":TIME,"WX":WX,"MAXT":MAXT,"MINT":MINT,"CI":CI,"PoP":PoP,"get_day":get_day}
df = pd.DataFrame(data)
df
```
